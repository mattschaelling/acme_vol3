# Problem 10.14

For the wages.csv dataset, start with the feature list ['female','educ','
exper','tenure','married','female*married','numdep','nonwhite'],
fit a linear model (including a constant!). Iteratively remove features based
on their 95% confidence interval and p-values (that is, remove the feature
with the highest p-value if its confidence interval contains 0, and rerun the
model) until all the features have $p < 0.05$. What happens to the adjusted
R-squared as you remove these features?

```python
df = pd.read_csv('wages.csv')
df['female*married'] = df['female']*df['married']
y = df['wage']
X = df[['female','educ','exper','tenure','married','female*married','numdep','nonwhite']]
X = sm.add_constant(X)
print(sm.OLS(y,X).fit().summary())
```

```python
X2 = df[['female','educ','exper','tenure','married','female*married','numdep']]
X2 = sm.add_constant(X2)
print(sm.OLS(y,X2).fit().summary())
```

```python
X3 = df[['female','educ','exper','tenure','married','female*married']]
X3 = sm.add_constant(X3)
print(sm.OLS(y,X3).fit().summary())
```

```python
X4 = df[['educ','exper','tenure','married','female*married']]
X4 = sm.add_constant(X4)
print(sm.OLS(y,X4).fit().summary())
```

```python
X5 = df[['educ','tenure','married','female*married']]
X5 = sm.add_constant(X5)
print(sm.OLS(y,X5).fit().summary())
```

The adjusted R-squared stays roughly constant. Very small increases, then returns to original value.

# Problem 10.15

For the wages.csv dataset, start with the feature list ['female','educ','
exper','tenure','married','female*married','numdep','nonwhite'],
fit a linear model (including a constant!), and compute the AIC and BIC for
that model. For every possible subset of features (but always including constants),
fit a linear model with those features and compute the AIC and the
BIC. Which combination of features gives the optimal (lowest) AIC? Which
gives the optimal BIC?

Hint: itertools.combinations may be useful here.

```python
for i in range(4):
    print('')
    print(i)
    for j in combinations(range(4),i+1):
        print(j)
```

```python
features = ['female','educ','exper','tenure','married','female*married','numdep','nonwhite']
models = []
AIC = []
BIC = []
y = df['wage']
for i in range(1,len(features)+1):
    print("\n--------------------------\nNumber of Features: {}\n".format(i))
    for feats in combinations(features, i):
        model = []
        for feat in feats:
            model.append(feat)
        models.append(model)
        X = df[model]
        X = sm.add_constant(X)
        results = sm.OLS(y,X).fit()
        print("Model:\t{}\nAIC:\t{}\nBIC:\t{}\n".format(model, results.aic, results.bic))
        AIC.append(results.aic)
        BIC.append(results.bic)
```

```python
opt_aic = np.argmin(AIC)
opt_bic = np.argmin(BIC)
print("Optimal AIC:\t{}\n\t\t{}\n\nOptimal BIC:\t{}\n\t\t{}".format(models[opt_aic],AIC[opt_aic], models[opt_bic], BIC[opt_bic]))
```
