# Problem 1
Apply the XGBoost classifier to the Fashion-MNIST dataset (https://github.com/zalandoresearch/fashion-mnist).  Do a grid search to find optimal values of parameters $\gamma$ and $\eta$ (the learning rate--I called this $\alpha$ in class). Compare your results to other classifiers, including random forests and SVC.

```python
# Load dataset.
X_train, y_train = mnist_reader.load_mnist('fashion_mnist_master/data/fashion', kind='train')
X_test, y_test = mnist_reader.load_mnist('fashion_mnist_master/data/fashion', kind='t10k')
```

```python
# Grid search for parameters.
'''
eta - learning rate [0,1], def=.3
gamma - minimum loss reduction to partition on a leaf [0,inf], def=0,
        more conservative as it gets larger
'''
boosted = xgb.XGBClassifier()
# param = {'eta':.3, 'gamma':0, 'silent':1}
Param_grid = {'learning_rate':np.linspace(.1,.7,4), 'gamma':np.linspace(0,1,4), 'silent':[1]}
models = ms.GridSearchCV(boosted, param_grid=Param_grid, n_jobs=-1)
models.fit(X_train, y_train)
results = models.score(X_test, y_test)
```

```python
# Compare results.
print(models.best_params_)
print(models.best_score_)
```
